```python
#14. Reverse a String Using a while Loop

s = input("Enter a string: ")
rev = ""
i = len(s) - 1

while i >= 0:
    rev += s[i]
    i -= 1

print("Reversed string:", rev)
```

```python
#15.Factorial using while loop
n = int(input("Enter a number: "))
fact = 1
i = 1

while i <= n:
    fact *= i
    i += 1

print("Factorial of", n, "is", fact)
```
